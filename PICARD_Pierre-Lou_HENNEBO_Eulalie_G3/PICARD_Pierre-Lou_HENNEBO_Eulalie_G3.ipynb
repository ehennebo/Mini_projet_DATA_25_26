#### Description du jeu de données : ####
Notre jeu de données est un fichier .asc (ASCII Grid). On peut utiliser genfromtxt pour charger les données. Sa taille est de ...

Il y a un header de 5 lignes de métadonnées essentiel à l'exploitation du set. 

Le MNT est composé d’une grille rectangulaire dont chaque noeud est doté d’une altitude. Chaque noeud est espacé par un pas de 5 m. (C’est un jeu de données 2D).
À un noeud de la grille correspond un numéro de colonne et un numéro de ligne (c,l), des coordonnées bidimensionnelles (X,Y) et une altitude Z exprimées en mètres (float précision 2CS après la virgule). Les coordonnées X et Y associées aux nœuds du MNT sont des valeurs métriques entières en projection Lambert 93.
Par exemple : si mon point de départ est de coordonnée x0 = 954997.5 m, le point suivant aura alors la coordonnée x1 = x0 + pas de la grille = 954997.5 + 5 = 955002.5 m. 
On construit comme cela un meshgrid python ! L’altitude Z = f(X,Y). 

ATTENTION à la correspondance des points : la première valeur d’altitude du jeu de donnée ne correspond pas à celle du point (0,0) mais à (0,Ymax) .
